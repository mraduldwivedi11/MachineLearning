```python
#########creating graph of x^3
import numpy as np
x=np.array([3,4,5,6])
y=x**3
plt.plot(x,y)
plt.show()

#############but we have to want a continuous graph
m=np.arange(2,5,0.1)
n=m**3
plt.plot(m,n,"r->")
plt.show()
```

# customizing the graphs

```python
m=np.arange(2,5,0.1)
n=m**3
plt.plot(m, n ,color="black", marker="o")   #########we can explicitly pass marker and color
plt.show()
```

```python
m=np.arange(2,5,0.1)
n=m**3
plt.plot(m, n ,color="black", linewidth=3)   #########line width to make width of line

plt.ylabel("x^3")

plt.xlabel("x")

plt.title("cubic graph")


#plt.axis([3,5,25,100])   ######## uses to define axis length


plt.show()
```

```python
#color, marker ,title ,label for axes
m=np.arange(2,5,0.1)
n=m**3
x=np.arange(2,6,0.3)
y=x**2

plt.plot(x,y ,color="red", linewidth=3, label="square")  ### here label means label provide for line  
plt.ylabel("x^2")
plt.xlabel("x")
plt.legend() ##to show that line label in seperate corner box
plt.title("square graph")

##   plt.axis([0,50,0,150])   ## this is use to change our axis custom 
                              ## argument has a list with 4 value first 2 are range of x axis 2nd two are range of y axis

    
plt.grid()  ### to plot grid

plt.text(2.5,60, "random text", fontsize=10)  ## to add text


plt.plot(m, n ,color="black", linewidth=3, label="cubic")  ### here label means label provide for line  
plt.ylabel("x^3")
plt.xlabel("x")
plt.legend()
plt.title("cubic graph")
plt.show()
```

# pie graph

```python
size=[26,12,25,37]

labels=["a","b","c","d"]    # to provide info for every part

colors=["red","yellow","blue","green"]  # to provide a color for every diff part

plt.title("pie graph exmple")

explode=[0.1,0,0,0]  ### this is use to cut a pie specially


plt.pie(size , labels=labels, colors=colors, autopct="%.2f%%", explode=explode , counterclock=False  , startangle=180)   ## autopct use for define % of each part

#####  counter colock use to decide how ur graph move  
##### start angle angle use for starting angle

plt.axis("equal")   ## for knowing what its for comment it and rub
plt.show()
```
