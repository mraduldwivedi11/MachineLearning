# dealing with NaN entries (removing NaN rows)

```python
df.iloc[1:3,1:3]=np.nan   ####her we create some nan values
print(df.head())
print(df.describe())
```

```python
#################  remove nan values
df.dropna(inplace=True)    ###to make permanent change set inplace=true
print(df.head())  # it removes row 1 and 2 
```

```python
##########  now we again set indexes
df.reset_index(drop=True , inplace= True)
print(df.head())
```

# adding values to NaN

```python
df.iloc[1:3,1:3]=np.nan 
df.head()
```

```python
df.sw.fillna(df.sw.mean(), inplace = True)
df.pl.fillna(df.pl.mean(), inplace = True)
df.head()
```
