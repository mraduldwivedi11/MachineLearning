```python
import numpy as np
import pandas as pd
from sklearn import datasets 
```

```python
def stepgradient(data, y , learning_rate , m , row , feature):
    m_slope=np.zeros(feature)
    l=len(data)
#     for j in range(len(m_slope)):
#         m_slope=(-2/l)*(y-m*x)*x[j]
    for j in range(len(m_slope)):   ## this loop we ran for m value of every feature
        for i in range(l):           ## this use for iterarte every row of a particular feature
            mul=0
            for k in range(14):
                mul+= m[k]*data[i,k]
            #m_slope[j]+= (-2/l)*( y[i]- (m[j]*data[i,j]))*data[i,j]
            m_slope[j]+= (-2/l)*( y[i]- mul)*data[i,j]
    for i in range(14):
        m[i]=m[i]-learning_rate*m_slope[i]
    return m


#     for i in range(l):
#         for j in range(len(m_slope)):
#             mul=0
#             for k in range(14):
#                 mul += m[k] * data[i,k]
#             m_slope[j]+= (-2/l)*(y[j]-mul)*data[i,j]


def gd(data, y , learning_rate , no_itr , row , feature):
    m=np.zeros(14)
    for i in range(no_itr):
        m=stepgradient(data, y, learning_rate, m,row , feature)
        print(cost(data,y,m))
    return(m)


```

```python
def cost(data,y,m):
    cost=0
    l=len(data)
    for i in range(l):
        mul=0
        for k in range(14):
            mul+= m[k]*data[i,k]
        cost += (1/l)*((y[i] - mul)**2)
    return (cost)
```

```python
def predict(data2,m):
    df=pd.DataFrame(data2)
    n=len(df.columns)
    df[n]=1
    x1=df.values
    y_pred=np.dot(x1,m)
   
    return y_pred
```

```python
def run():
    test_data=np.loadtxt("0000000000002417_test_boston_x_test.csv" , delimiter=",")
    old_data=np.loadtxt("0000000000002417_training_boston_x_y_train.csv" , delimiter=",")
   # olddata = np.loadtxt("data.csv", delimiter=",")
    y=old_data[:,13]   # remove the oytput column
    row=len(y)
       # add one for intercept column
    x=old_data[:,0:13]
    df=pd.DataFrame(x)
    df["13"]=1
    data=df.values
    # data=np.append(olddata[:,0] , z , axis=1)
    
    feature=14    #### this variable decides size of araay of m
    learning_rate=0.1
    no_itr=150
    m=gd(data , y , learning_rate , no_itr , row , feature)
    #print(m)
    y_pred=predict(test_data,m)
    np.savetxt("predictions1.csv",y_pred,delimiter=",")
```
