```python
import numpy as np
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from pprint import pprint
```

```python
from sklearn.datasets import fetch_20newsgroups
```

```python
twenty_train = fetch_20newsgroups(subset="train", shuffle=True)
```

```python
dictionary={}
sw=stopwords.words('english')
shortword = re.compile(r'\W*\b\w{1,3}\b')
# for i in range(0,10):
# for i in range(0,len(twenty_train['data'])+1):
for i in range(0,20):
    dictionary[i]={}
    
    
#for i in range(0,10):
for i in range(0,len(twenty_train['data'])):
    strn=re.sub(r'[^\w\s]','',twenty_train['data'][i])
    strn=re.sub(r'\w*\d\w*', '', strn).strip()
    shortword.sub('', strn)
    p=twenty_train['target'][i]
    word_tokens = word_tokenize(strn)
    word_tokens = [word.lower() for word in word_tokens]
    pure_features=[]
    for w in word_tokens:
        if w.isalpha():
            if w not in sw:
                if len(w)>2:
                    if w in dictionary[p]:
                        dictionary[p][w]=dictionary[p][w]+1
                    else:
                        dictionary[p][w]=1
                    pure_features.append(w)
#     print(pure_features)
#     print("----------------------")
# print(dictionary[4])
```

```python

import operator
for i in range(0,20):
    dictionary[i] = sorted(dictionary[i].items(), key=operator.itemgetter(1),reverse=True)[0:1000]
# print(type(sorted_d))
# pprint(sorted_d[0:1000])
```
