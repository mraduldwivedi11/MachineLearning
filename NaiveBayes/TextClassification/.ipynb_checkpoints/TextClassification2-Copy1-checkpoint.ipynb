```python
import nltk
nltk.download('stopwords')
import numpy as np
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from pprint import pprint
import operator
```

```python

from sklearn.datasets import fetch_20newsgroups
twenty_train = fetch_20newsgroups(subset="train", shuffle=True)
```

```python
dictionary={}
sw=stopwords.words('english')
shortword = re.compile(r'\W*\b\w{1,3}\b')

for i in range(0,len(twenty_train['data'])):
    strn=re.sub(r'[^\w\s]','',twenty_train['data'][i])
    strn=re.sub(r'\w*\d\w*', '', strn).strip()
    shortword.sub('', strn)
    word_tokens = word_tokenize(strn)
    word_tokens = [word.lower() for word in word_tokens]
    for w in word_tokens:
        if w.isalpha():
            if w not in sw:
                if len(w)>2:
                    if w in dictionary:
                        dictionary[w]=dictionary[w]+1
                    else:
                        dictionary[w]=1
```

```python
pprint(dictionary)
```

```python
features = sorted(dictionary.items(), key=operator.itemgetter(1),reverse=True)[0:2000]
features_list=[]
for i in features:
    features_list.append(i[0])
pprint(features_list)
```

```python
import pandas as pd
# df=pd.DataFrame(rows=len(twenty_train['data']),columns=features_list)
#print(df[0]['subject'])

df = pd.DataFrame(0, index=range(0,len(twenty_train['data'])), columns=range(0,2000))
         
for i in range(0,len(twenty_train['data'])):
        word_tokens = word_tokenize(twenty_train['data'][i])
        word_tokens = [word.lower() for word in word_tokens]
        for word in word_tokens:
            if word in features_list:
                column_no=features_list.index(word)
                df.iloc[i,column_no]=df.iloc[i,column_no]+1
            else:
                continue
```

```python
print(len(twenty_train['target']))
df.shape
twenty_train['target'][0]
```

```python
x=df.values
```

```python
from sklearn import model_selection
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(x,twenty_train['target'],test_size=0.25,random_state=0)
```

```python

from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(X_train, Y_train)
```

```python
Y_pred=clf.predict(X_test)
```

```python
single_predict=clf.predict(X_test[52].reshape(1, -1))
print(type(single_predict))
print(twenty_train['target_names'][single_predict[0]])
```
