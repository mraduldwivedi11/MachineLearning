```python
import numpy as np
```

```python
def stepgradient(data,learning_rate,m,c):
    m_slope=0
    c_slope=0
    l=len(data)
    for i in range(l):
        x=data[i,0]
        y=data[i,1]
        m_slope+= (-2/l)*(y-m*x-c)*x
        c_slope+= (-2/l)*(y-m*x-c)
    m_new = m- learning_rate*m_slope
    c_new = m-learning_rate*c_slope
    return(m_new,c_new)
```

```python
def gd(data,learning_rate,no_itr):
    m=0
    c=0
    for i in range(no_itr):
        m,c=stepgradient(data,learning_rate,m,c)
        print(cost(data,m,c))
    return(m,c)
```

```python
def cost(data,m,c):
    cost=0
    l=len(data)
    for i in range(l):
        x=data[i,0]
        y=data[i,1]
        cost += (1/l)*((y-m*x-c)**2)
    return cost
```

```python
def run():
    data = np.loadtxt("data.csv", delimiter=",")
    learning_rate=0.0001
    no_itr=100
    m,c=gd(data,learning_rate,no_itr)
    print(m,c)
```
